Basis der Datenbeschaffung ist die Spotify API. Diese liefert neben Metadaten der Songs (Name, Künstler etc.) ebenfalls die zu untersuchenden Soundeigenschaften.
Dazu werden primär folgende Endpoints verwendet:

| Endpoint | Beschreibung |
|------|------|
| https://api.spotify.com/v1/tracks/ | Liefert Metainformationen zu einem Song anhand dessen Spotify-ID|
| https://api.spotify.com/v1/search?include_external=audio | Liefert Songs, Alben, Künstler oder Genres basierend auf einer dynamischen Query (wird benötigt, um die Spotify-ID anhand eines Songnamens zu ermitteln) |
| https://api.spotify.com/v1/audio-features | Liefert Soundeigenschaften zu einem mitgegebenen Song anhand dessen Spotify-ID |
| https://api.spotify.com/v1/audio-analysis/ | Liefert Soundanalysen (z.B. Anzahl Beats, Bars etc.) zu einem mitgegebenen Song anhand dessen Spotify-ID |

Von den Rückgabewerten der `/audio-features` und `/audio-analysis` Endpoints lassen sich für die Arbeit spannende Soundeigenschaften verwenden:

| Soundeigenschaft | Beschreibung |
|------|------|
| acousticness | Confidence-Index (0-1), ob ein Song akustisch ist oder nicht |
danceability | Index (0-1), wie gut zu dem Song getanzt werden kann |
duration_ms | Songlänge in Millisekunden |
energy | Index (0-1), wie energetisch ein Song ist (höherer Wert bedeutet grössere Intensität und Aktivität) |
instrumentalness | Confidence-Index (0-1), ob der Song ein reines Instrumental ist (je geringer der Wert, desto eher ist Gesang vorhanden) |
liveness | Confidence-Index (0-1), ob der Song live aufgenommen wurde
loudness | Durchschnittliche Lautstärke in Dezibel (db)
speechiness | Confidence-Index (0-1), ob der Song primär Sprechgesang enthält
tempo | Beats pro Minute (BPM)
valence | Index (0-1), ob ein Song eher positive Gefühle (hoher Wert) auslöst oder negative (niedriger Wert)

### Spotify-Chartdaten

Spotify misst die Chartplatzierungen von Songs durch Plays in Relation zu wöchentlichen Zeiträumen, also “charten” jeweils immer die 200 Songs mit den meisten Plays pro Woche, entweder global (ohne Herkunftsfilter für die Plays) oder länderbasiert.
Um ein gesamteinheitliches Ranking der Anzahl Plays für ein ganzes Jahr zu bekommen, sollen also die wöchentlichen Plays jeder Woche addiert werden, um so eine finale Liste der populärsten Songs zu bekommen.
Die Anzahl der Songs kann somit auch je nach Land variieren, da sich in bestimmten Regionen bestimmte Songs potenziell für längere Zeit in den wöchentlichen Top 200 halten können als in anderen. Das Problem ist, dass die Wochencharts nicht über einen offiziellen API-Endpoint abgeholt werden können, sondern über die Charts-API, welche nur über ein Frontend abgerufen werden soll (Spotify bietet keine Dokumentation zur Authentifizierung mit der Charts API).
Trotzdem kann man dies automatisch lösen, indem man sich einfachheitshalber über das Frontend einmal authentifiziert und die entsprechenden Header dann dem GET-Request in Python mitgibt.


````python
# visit https://charts.spotify.com/charts/view/regional-gb-weekly/2020-12-31 (or any other week/region) using Chrome, open network tab, refresh the site, search for a 200 GET on https://charts-spotify-com-service.spotify.com
# rightlick the request, choose copy -> as cURL (bash), convert the cURL to a py.requests request using https://curlconverter.com/, and paste the header here
headers = {}
````

````python
# rest of the URL will be dynamic
base_url = "https://charts-spotify-com-service.spotify.com/auth/v0/charts/"
# the countries and years we want to fetch, multilist for the years as we also need to know the start date for the very first weekly chart of each year (obtained manually through spotifys frontend)
countries = ['us', 'ch', 'de', 'gb']
spotify_year_start = [['2020', '2020-01-09'], ['2021', '2021-01-07'], ['2022', '2022-01-06']]
full_track_data = {}

# we iterate through each year and first obtain all week-charts for this year (using timedelta)
for start in spotify_year_start:
  full_track_data[start[0]] = {}
  start_date = datetime.datetime.strptime(start[1], '%Y-%m-%d')
  end_date = datetime.datetime.strptime(start[0] + '-12-31', '%Y-%m-%d')
  dates = []
  current_date = start_date
  while current_date <= end_date:
    dates.append(current_date.strftime('%Y-%m-%d'))
    current_date += datetime.timedelta(days=7)

  # once we have all weeks of said year, we iterate through the countries and build the final API request URL using the combination of each country and each week of the year
  for country in countries:
    full_track_data[start[0]][country] = {}
    for date in dates:
      request_url = base_url + 'regional-' + country + '-weekly/' + date
      # once we have built our url, we send out a request using the previously prepared headers (authorized manually in frontend)
      response = requests.get(request_url, headers=headers).json()
      # the response will be JSON serving all 200 song entires of said week which we start iterating through
      if 'entries' not in response:
      for entry in response['entries']:
          track_id = entry['trackMetadata']['trackUri']
          track_plays = int(entry['chartEntryData']['rankingMetric']['value'])
          track_name = entry['trackMetadata']['trackName']
          # once we have obtained each spotify-ID and its amount of plays, we check whether the song has already been added to our main dictionary (by another week/loop), if not we fully add it, if yes we just sum up the total plays
          if track_id in full_track_data[start[0]][country]:
            full_track_data[start[0]][country][track_id]['plays'] += track_plays
          else:
            full_track_data[start[0]][country][track_id] = {}
            full_track_data[start[0]][country][track_id]['plays'] = track_plays
            full_track_data[start[0]][country][track_id]['name'] = track_name
            full_track_data[start[0]][country][track_id]['artists'] = []
            for artist in entry['trackMetadata']['artists']:
              full_track_data[start[0]][country][track_id]['artists'].append({'artist_name': artist['name'], 'artist_id': artist['spotifyUri']})
````

Nun sind sämtliche Songs pro Land / Jahr abgerufen, es gilt nun diese anhand ihrer Plays zu sortieren, sodass der Song mit den meisten Plays die Chart Nummer #1 hat.

````python
for year, countries in full_track_data.items():
    for country, tracks in countries.items():
        # sort the tracks by plays in descending order
        sorted_tracks = sorted(tracks.items(), key=lambda x: x[1]['plays'], reverse=True)
        for i, (track_id, track_data) in enumerate(sorted_tracks):
            # assign placement to each track based on their plays
            track_data['placement'] = i+1
        # replace the unranked track data with the ranked data
        full_track_data[year][country] = dict(sorted_tracks)
````

*Da* wir nun alle Daten der selbstgebastelten "Spotify Charts" haben, sollen diese in die Datenbank abgefüllt werden, dazu müssen in verschiedenen Jahren / Ländern doppelt vorkommende Songs nocheinmal zusammengefügt werden und schlussendlich nur noch durch ihren `placement` wert in der `chart_placement` differenziert werden.

````python
for year, countries in full_track_data.items():
  for country, tracks in countries.items():
    # fetch ID of the chart thats currently being iterated through
    cursor.execute('SELECT chart_id FROM chart WHERE name=? AND measured_year=? AND country=?', ('spotify', year, country))
    chart_id = cursor.fetchone()
    for track_id, track_data in tracks.items():
      # remove spotify:track prefix within the track_id (not needed for further API calls)
      cleaned_track_id = track_id.replace('spotify:track:', '')
      track_name = track_data['name']
      # check if the track already exists in the database
      cursor.execute('SELECT track_id FROM track WHERE track_id=?', (cleaned_track_id, ))
      row = cursor.fetchone()
      # if not we add it
      if row is None:
        cursor.execute('INSERT INTO track (track_id, name) VALUES (?, ?)', (cleaned_track_id, track_name))
        # if it doesnt exist, we also need to link all artists to it
        for artist in track_data['artists']:
          artist_id = artist['artist_id'].replace('spotify:artist:', '')
          artist_name = artist['artist_name']
          cursor.execute('SELECT artist_id FROM artist WHERE artist_id=?', (artist_id,))
          artist_row = cursor.fetchone()
          # if these dont exist either, we need to add them first
          if artist_row is None:
            cursor.execute('INSERT INTO artist (artist_id, name) VALUES (?, ?)', (artist_id, artist_name))
          cursor.execute('INSERT INTO artist_track (artist_id, track_id) VALUES (?, ?)', (artist_id, cleaned_track_id))
      # finally we link the track (wheter it existed or was just added) to the chart thats been currently iterated through
      cursor.execute('INSERT INTO chart_placement (track_id, chart_id, placement, spotify_plays) VALUES (?, ?, ?, ?)', (cleaned_track_id, chart_id[0], track_data['placement'], track_data['plays']))
````

1.   Durch Besuch der Webseite kann erkannt werden, dass die Daten über verschiedene URLs in einem immer gleich bleibenden HTML Format angezeigt werden
2.   Nach Analyse der URLs zeigt sich, dass die Länder im Teil ```/single-charts-XX/``` bestimmt sind und das Jahr im Folgeteil ```/hits-XXXX/```, letztlich beinhalten alle Charts mehrere Seiten, welche im letzten Teil der URL bestimmt sind ```/X/```
3.   Nach Analyse der HTML-Struktur zeigt sich, dass die Einträge immer in einem `<div>` Element mit der ID `dataList` sind
4.   Dank dieser Strukur kann mittels `requests` die einzelnen URLs abgerufen werden und anschliessend mit `BeatifulSoup` der HTML Code auf relevante Elemente heruntergebrochen und extrahiert werden

````python
# prepare url data specifically for chartsurfer
base_url = 'https://www.chartsurfer.de/musik/single-charts-'
chartsurfer_data = [{'suffix': '/top-75', 'country': 'gb', 'country_url': 'uk'}, {'suffix': '/top-100', 'country': 'ch', 'country_url': 'schweiz'}, {'suffix': '/top-100', 'country': 'us', 'country_url': 'usa'}, {'suffix': '/top-100', 'country': 'de', 'country_url': 'deutschland'}]
chartsurfer_charts = {}

# simple indexed loop to iterate through year end numbers
for x in range(3):
  year = '202' + str(x)
  chartsurfer_charts[year] = {}
  for country_data in chartsurfer_data:
    chartsurfer_charts[year][country_data['country']] = []

    # simple indexed loop to increase a page index, as each country/years charts are split dynamically to up to 5 pages
    for page in range(1, 5):
      url = base_url + country_data['country_url'] + '/jahrescharts/hits-' + year + country_data['suffix'] + '/' + str(page)
      data = requests.get(url)

      # get the sites html
      html = BeautifulSoup(data.text, 'html.parser')
      # find the table div containing chart data and each chart entry
      chart_table = html.find('div', {'id': 'dataList'})

      # if no charts were found on the requested page, we have likely exceeded the last page number and will break the page loop
      if chart_table is None:
        break

      # get all chart entries, these have the same css class
      entries = chart_table.findAll('div', {'class': 'position-relative mt-2'})

      # iterate through all chart entries, extracting track name, artist name and placement
      for entry in entries:
        placement = entry.find('div', {'class': 'place'}).get_text()
        track_data = entry.select("a[href*=artist]")
        track_name = track_data[0].get_text()
        track_artist = track_data[1].get_text()
        chartsurfer_charts[year][country_data['country']].append({'placement': placement, 'name': track_name, 'artist': track_artist})
````

Im Folgeschritt sollen für die Songs der Ländercharts die Spotify ID ermittelt werden, diese ist relevant um zukünftig die Soundeigenschaften abholen zu können.

````python
# loop through chartsurfer_charts
for year, countries in chartsurfer_charts.items():
  for country, tracks in countries.items():
    for track in tracks:
      # cleanup song name / arist, so that the song can be found more accurately via spotify (spotify does not use generic 'feat' in song titles but rather combines multiple aritsts through tagging)
      track_artist = track['artist'].replace('feat.', '').replace('Feat.', '').replace(' x ', ' ').replace('&', '').replace("'", "").replace(' X ', ' ')
      track_name = track['name'].replace("'", "")
      track['spotify_data'] = {}
      # send a search request to spotify
      spotify_track_data = spotify.search(q='track:' + track_name + ' artist:' + track_artist, type='track', limit=1, market=country)['tracks']
      # if a song has been found, parse its information and write it back to the chartsurfer dictionary
      if len(spotify_track_data['items']) != 0:
        track['spotify_data']['name'] = spotify_track_data['items'][0]['name']
        track['spotify_data']['id'] = spotify_track_data['items'][0]['uri']
        track['spotify_data']['artists'] = []
        for artist in spotify_track_data['items'][0]['artists']:
          track['spotify_data']['artists'].append({'artist_name': artist['name'], 'artist_id': artist['id']})
````

Nach Betrachtung der Resultate wurde erkannt, dass bestimmte Songs redundant aber dennoch mit verschiedenen `track_id` gespeichert wurden. Dies kann daran liegen, dass die Songs auf verschiedenen Alben erschienen sind oder im Verlaufe der Zeit neu veröffentlicht wurden (aus Copyright- oder anderen Gründen). Dies wird nun bereinigt, indem zuerst der `track_name` auf redundanz gepfrüft wird, trifft dies für mehrere `track_id` zu, so wird geprüft, ob diese ebenfalls vom selben Künstler stammen, trifft beides zu - also gleicher Song vom gleichen Künstler - werden die redundanten `track_id` von der Datenbank gelöscht und allfällige Chart-Platzierungen mit einer einzigen `track_id` versehen.

````python
from collections import defaultdict

# we query to get master and redundant track_ids with the same name in track table and same artist_id in artist_track table
query = '''
SELECT t1.track_id AS master_track_id, t2.track_id AS redundant_track_id
FROM track AS t1
JOIN artist_track AS at1 ON t1.track_id = at1.track_id
JOIN track AS t2 ON t1.name = t2.name
JOIN artist_track AS at2 ON t2.track_id = at2.track_id AND at1.artist_id = at2.artist_id
WHERE t1.track_id < t2.track_id
'''

# fetch the results
cursor.execute(query)
results = cursor.fetchall()

# then construct the desired output format
output = defaultdict(list)
for master_track_id, redundant_track_id in results:
    output[master_track_id].append(redundant_track_id)
````

````python
def update_chart_placement_and_delete_artist_track_audio_feature(conn, master_track_id, redundant_track_ids):
    cursor = conn.cursor()
    for redundant_track_id in redundant_track_ids:
        # update track_id in chart_placement table
        cursor.execute('''
            UPDATE chart_placement
            SET track_id = ?
            WHERE track_id = ?
        ''', (master_track_id, redundant_track_id))

        # delete rows in artist_track table with redundant track_id
        cursor.execute('''
            DELETE FROM artist_track
            WHERE track_id = ?
        ''', (redundant_track_id,))

        cursor.execute('''
            DELETE FROM audio_feature
            WHERE track_id = ?
        ''', (redundant_track_id,))

        cursor.execute('''
            DELETE FROM track
            WHERE track_id = ?
        ''', (redundant_track_id,))

    conn.commit()

# iterate through the output dictionary and perform the required deletion declared just above
for master_track_id, redundant_track_ids in output.items():
    update_chart_placement_and_delete_artist_track_audio_feature(connection, master_track_id, redundant_track_ids)

connection.commit()
connection.close()
````

##Hypothese 5: Soundeigenschaften unterscheiden sich zwischen Ländern und zwischen Charts

````python
import sqlite3

# Verbindung zur Datenbank herstellen
conn = sqlite3.connect('bina.db')

# SQL-Abfrage zur Berechnung der durchschnittlichen Soundeigenschaften pro Land und Chartart für die Top 50
query = """
SELECT c.country, c.name AS chart_name,
       AVG(af.duration_ms) AS avg_duration_ms,
       AVG(af.acousticness) AS avg_acousticness,
       AVG(af.danceability) AS avg_danceability,
       AVG(af.energy) AS avg_energy,
       AVG(af.instrumentalness) AS avg_instrumentalness,
       AVG(af.liveness) AS avg_liveness,
       AVG(af.loudness) AS avg_loudness,
       AVG(af.speechiness) AS avg_speechiness,
       AVG(af.tempo) AS avg_tempo,
       AVG(af.valence) AS avg_valence
FROM chart AS c
INNER JOIN chart_placement AS cp ON c.chart_id = cp.chart_id
INNER JOIN audio_feature AS af ON cp.track_id = af.track_id
WHERE cp.placement <= 50
GROUP BY c.country, c.name
"""

# Abfrage ausführen
cursor = conn.execute(query)

# Ergebnisse abrufen und in Variablen speichern
results = cursor.fetchall()
averages = []
for row in results:
    country = row[0]
    chart_name = row[1]
    avg_duration_ms = row[2]
    avg_acousticness = row[3]
    avg_danceability = row[4]
    avg_energy = row[5]
    avg_instrumentalness = row[6]
    avg_liveness = row[7]
    avg_loudness = row[8]
    avg_speechiness = row[9]
    avg_tempo = row[10]
    avg_valence = row[11]

    averages.append({
        'country': country,
        'chart_name': chart_name,
        'avg_duration_ms': avg_duration_ms,
        'avg_acousticness': avg_acousticness,
        'avg_danceability': avg_danceability,
        'avg_energy': avg_energy,
        'avg_instrumentalness': avg_instrumentalness,
        'avg_liveness': avg_liveness,
        'avg_loudness': avg_loudness,
        'avg_speechiness': avg_speechiness,
        'avg_tempo': avg_tempo,
        'avg_valence': avg_valence
    })

# Verbindung zur Datenbank schließen
conn.close()

# Ausgabe der durchschnittlichen Soundeigenschaften pro Land und Chartart
for avg in averages:
    print(f"Land: {avg['country']}, Chart: {avg['chart_name']}")
    print(f"Durchschnittliche Soundeigenschaften:")
    print(f"Dauer (ms): {avg['avg_duration_ms']}")
    print(f"Akustik: {avg['avg_acousticness']}")
    print(f"Tanzbarkeit: {avg['avg_danceability']}")
    print(f"Energie: {avg['avg_energy']}")
    print(f"Instrumentalität: {avg['avg_instrumentalness']}")
    print(f"Lebendigkeit: {avg['avg_liveness']}")
    print(f"Lautstärke: {avg['avg_loudness']}")
    print(f"Sprechanteil: {avg['avg_speechiness']}")
    print(f"Tempo: {avg['avg_tempo']}")
    print(f"Gefühl: {avg['avg_valence']}")
    print()
````

Die Auswertung der Durchschnittlichen Soundeigenschaften pro Land und Chartart weist auf den ersten Blick keine grossen Unterschiede auf. Die Durchschnittswerte befinden sich grundsätzlich auf ähnlichem Niveau.


````python
import matplotlib.pyplot as plt
import numpy as np

# Daten für das Diagramm vorbereiten
properties = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence']
chart_names = np.unique([avg['chart_name'] for avg in averages])
avg_values = {prop: [] for prop in properties}
for avg in averages:
    for prop in properties:
        avg_values[prop].append(avg[f'avg_{prop}'])

# Balkendiagramm erstellen
plt.figure(figsize=(12, 6))
bar_width = 0.2
index = np.arange(len(properties))

for i, chart_name in enumerate(chart_names):
    plt.bar(index + i * bar_width, [avg_values[prop][i] for prop in properties], bar_width, label=chart_name)

plt.xlabel('Soundeigenschaft')
plt.ylabel('Durchschnittswert')
plt.title('Vergleich der Soundeigenschaften pro Chartart')
plt.xticks(index + bar_width * (len(chart_names) - 1) / 2, properties)
plt.legend()
plt.tight_layout()
plt.show()
````

Es gibt keine signifikanten Unterschiede zwischen den Soundeigenschaften in den lokalen und Spotify-Charts. In der Tendenz haben Songs in den lokalen Charts mehr Energie und Danceability, dafür weniger Acousticness.

````python
import matplotlib.pyplot as plt

# Liste der Soundeigenschaften
properties = ['danceability', 'energy', 'liveness', 'valence']

# Daten für das Diagramm vorbereiten
countries = [avg['country'] for avg in averages]
chart_names = [avg['chart_name'] for avg in averages]
avg_values = {prop: [avg[f'avg_{prop}'] for avg in averages] for prop in properties}

# Balkendiagramme erstellen
fig, axs = plt.subplots(len(properties), 1, figsize=(10, 8), sharex=True)

for i, prop in enumerate(properties):
    axs[i].bar(range(len(averages)), avg_values[prop])
    axs[i].set_ylabel(f'Ø {prop.capitalize()}')
    axs[i].set_title(f'Durchschnittliche {prop.capitalize()} pro Land und Chartart')

# X-Achse beschriften
plt.xticks(range(len(averages)), zip(countries, chart_names), rotation='vertical')
plt.xlabel('Land und Chartart')

plt.tight_layout()
plt.show()
````
