#### 2.2.2.1 Wetterdaten

Da, wie erwartet, aus den Daten des FC St. Gallen keine Wetter-Informationen am Spieltag abgeleitet werden können, müssen diese über Internet-Quellen besorgt werden.

Die Wetterdaten am Spieltag werden über eine API abgerufen. Die API wird von Open-Meteo zur Verfügung gestellt und bietet unterschiedlichste abrufbare Variablen zum Abruf an, wie in der nachfolgenden Grafik ersichtlich ist (https://open-meteo.com/en/docs/historical-weather-api).


Wetter API.png

**Abbildung Wetterdaten von Open-Meteo (www.open-meteo.com)**

Abgerufen werden die Daten über nachfolgende API URL:
«https://archive-api.open-mete-o.com/v1/archive?latitude={Variable}&longitude={Variable}&start_date={Variable}&end_date={Variable}&hourly=temperature_2m&min={Variable}&max={Variable}»
Diese API URL wird je nach benötigten Wetter-Variablen noch ergänzt oder gekürzt. Überall wo "{Variable}" in dem URL steht, wird ein Parameter/eine Variable benötigt für eine korrekte Abfrage. Für die spätere Analyse gelten nachfolgende Parameter.


| Bezeichnung               | Variable/Parameter       |
| ------------------------- | ------------------------ |
| Latitude kybunpark                 |  47.408327      |
| Longitude kybunpark                |  9.305620       |
| Start date                | Abhängig von Spieltag    |
| End date                  | Abhängig von Spieltag    |
| Hourly Weather Variables  | Noch zu definieren in der Datenanalyse |


Wie in der Tabelle erläutert, wird während der späteren Analyse genauer definiert, welche Variablen konkret abgefragt werden.

**Wichtig:**

Die Daten können nur auf stündlicher Basis abgerufen werden. Es können somit keine genauen Abrufe auf Minutenbasis ausgeführt werden. Z.B. hat dies zur Folge, dass bei Informationsabfrage zum Zeitpunkt eines Anpfiffs (z.B. 16:45Uhr) auf die volle Stunde 17Uhr abgefragt werden muss.


#### Funktionen für den Wetterdaten-Abrufen

Für die Analyse wurden zwei Funktionen erstellt, über welche die Wetterdaten für die Stadionlocation und für die gegebene Anpfiffszeit (Datum und Uhrzeit) abgerufen werden kann. Anhand dieser Funktionen soll zu späteren Zeitpunkten Informationen über Niederschlag und Temperatur abrufbar sein.

ToDo: Aleks --> Text kontrollieren und ergänzen, ggf. in anderem Kapitel platzieren.

```python
# Liefert die Wetterdaten für die Stadionlocation (Niederschlag und Temperatur)
def fetch_weather_data(start_date):
    latitude = 47.408327
    longitude = 9.305620
    end_date = start_date
    weather_info = 'temperature_2m,precipitation,rain,snowfall'
    url = f"https://archive-api.open-meteo.com/v1/archive?latitude={latitude}&longitude={longitude}&start_date={start_date}&end_date={end_date}&hourly={weather_info}&timezone=auto"
    response = requests.get(url).json()
    data = response['hourly']
    data['time'] = pd.to_datetime(data['time'])
    df = pd.DataFrame(data)
    return df
```

```python
# Liefert die Wetterdaten für die Anpfiffszeit
def fetch_weather_data_for_match(start_date, start_time):
    df_weather_for_matchday = fetch_weather_data(start_date)

    # Konvertiere 'start_time' zu datetime.time Objekt und rundet zur nächsten Stunde
    match_hour = pd.to_datetime(start_time).round('H').hour

    # Jetzt filtern wir die Wetterdaten für die spezifische Stunde des Anpfiffs
    df_weather_for_matchhour = df_weather_for_matchday[df_weather_for_matchday['time'].dt.hour == match_hour]

    return df_weather_for_matchhour
```

## 2.4 Datenmodellierung

Datenmodellierung ist ein zentraler Aspekt von Business Analytics, bei dem Daten aus verschiedenen Quellen zusammengeführt und in einer strukturierten Form organisiert werden. Durch die Definition von Beziehungen und Attributen wird ein logisches Modell geschaffen, das die Grundlage für effektive Datenanalyse und Berichterstattung bildet. Basis der nachfolgenden Analyse bildet die Datei des FC St. Gallen mit den Ticketverkäufen. In diesem Kapitel werden vor allem Informationen aus den weiteren Tabellen der Haupttabelle über die Ticketverkäufe angehängt.

### 2.4.1 Aufbereitung der Masterdatei 'fcsg_tickets_master_data'

Wie zuvor angeschprochen, dient die Tabelle der FC St. Gallen über die Ticketverkäufe als "Master" und wird mit weiteren benötigten Informationen angereichert. Zudem werden teils inhaltliche Informationen ummodelliert.

#### 2.4.1.1 Ummodellierung der Spalte 'Anfpfiff'

Die Spalte 'Anpfiff' hat das Format DD.MM.YYYY HH:MM:SS. Diese Informationen werden auseinandergezogen und in separaten Spalten ergänzt. Es werden die neuen Spalte "Anpfiff Datum", "Anpfiff Zeit" und "Anpfiff Wochentag" erstellt und aus der Spalte 'Anpfiff' abgeleitet. Nachfolgend den Code für diese ergänzende Modellierung.

```python
# Konvertierung von 'Anpfiff' und 'Änderungsdatum' zu datetime
fcsg_ticket_master_data['Anpfiff'] = pd.to_datetime(fcsg_ticket_master_data['Anpfiff'])
fcsg_ticket_master_data['Änderungsdatum'] = pd.to_datetime(fcsg_ticket_master_data['Änderungsdatum'])

# Erstellung von 'Anpfiff Datum', 'Anpfiff Zeit' und 'Anpfiff Wochentag' Spalten
fcsg_ticket_master_data['Anpfiff Datum'] = fcsg_ticket_master_data['Anpfiff'].dt.date
fcsg_ticket_master_data['Anpfiff Zeit'] = fcsg_ticket_master_data['Anpfiff'].dt.time
fcsg_ticket_master_data['Anpfiff Wochentag'] = fcsg_ticket_master_data['Anpfiff'].dt.day_name()
```

#### 2.4.1.2 Ergänzung der Saison sowie des Spieltages in der Liga

Weiters werden in neuen Spalten für jeden Ticketverkauf die entsprechende Saison in der Spalte 'Saison' und der entsprechende Spieltag in der Spalte 'Spieltag' ergänzt. Die Daten über Saison und Spieltag kommen aus der Tabelle 'spieltag_master_data'. Basierend auf dem 'Anpfiff Datum' werden die zugehörigen Informationen zu Saison und Spieltag der Masterdatei angehängt. Nachfolgend den Code dazu.

```python
# Konventierung von 'Spieldatum FC St. Gallen' zu datetime
spieltag_master_data['Spieldatum FC St. Gallen'] = pd.to_datetime(spieltag_master_data['Spieldatum FC St. Gallen']).dt.date

# Merge zwischen spieltag und master data
fcsg_ticket_master_data = fcsg_ticket_master_data.merge(spieltag_master_data,
                                            left_on='Anpfiff Datum',
                                            right_on='Spieldatum FC St. Gallen',
                                            how='inner')

# Spalte 'Spieldatum FC St. Gallen' entfernen
fcsg_ticket_master_data = fcsg_ticket_master_data.drop('Spieldatum FC St. Gallen', axis=1)
```

#### 2.4.1.3 Ergänzung von 'Tage bis zum Spiel'

Eine noch fehlende Information ist die, wie viel Tage vor (oder auch nach) dem Spiel die Tickets gekauft wurden. In der Masterdatei über die Ticketverkäufe besteht die Information 'Anpfiff' mit dem Inhalt Datum und Zeit des Spielbeginns sowie die Information 'Änderungsdatum' mit dem Zeitstempel der Bezahlung der Tickets. Es wird der Masterdatei eine neue Spalte angehäng und darin über die Formel Anpfiff - Änderungsdatum die Anzahl Tage zwischen Spielanpfiff und Bezahlung des Kaufauftrags berechnet. Nachfolgend ist der entsprechende Code für die Umsetzung.

**Wichtig**

Wie schon weiter oben im Kapitel 2.2.1.4 erläutert wird, ist die Information in der Spalte 'Änderungsdatum' das Datum, an dem die Tickets bezahlt wurden. Bei Ticketkäufen an der Tageskasse ist dies oft gleich wie das Datum des Ticketbezuges. In Anderen Fällen kann das Ticket jedoch auch zu einem späteren Zeitpunkt abgebucht werden und somit stimmt das 'Änderungsdatum' nicht mit dem Datum des Ticketbezuges überein. Dies verzerrt die Analyse. Bei Analysen mit der Variable 'Tage bis zum Spiel' wird jeweils darauf hingewiesen.

```python
# Berechnen Sie die Anzahl der Tage bis zum Spiel
fcsg_ticket_master_data['Tage bis zum Spiel'] = (fcsg_ticket_master_data['Anpfiff'] - fcsg_ticket_master_data['Änderungsdatum']).dt.days
```

#### 3.1.2 Analyse 2: Zeitlicher Ablauf der Ticketkaufe vor dem Spiel

Als weitere Analyse im allgemeinen Themenbereich wird der zeitliche Ablauf der Ticketkäufe genauer betrachtet. Wie viel Tage vor dem Spiel werden wie viel Tickets verkauft und was sind Verkaufsarten bei den Peaks? Hierzu werden mehrere Auswertungen erstellt, um unterschiedliche Blickwinkel zu erhalten.


In der Masterdatei gibt es total 73 verschiedene Verkaufsarten. Damit eine vereinfachte Analyse durchgeführt werden kann, werden diese in Gruppen unterteilt wie im Kapitel 2.4 Datenmodellierung ersichtlich wurde. Nachfolgend wird genauer analysiert, in welchen Verkaufsgruppen die Anzahl an Ticketverkäufen steigt. Betrachtet wird dabei die Anzahl Tage vor dem Spiel. **Wichtig**

In der Spalte 'Tage bis zum Spiel' gibt es auch negative Zahlen. Diese ergebene sich, wie weiter oben schon erläutert, wenn die Tickets nach dem Spiel bezahlt wurden und dadurch das 'Änderungsdatum' in der Mastertabelle nach dem 'Anpfiff' ist. ### 3.2.1 Analyse 1: Anzahl Ticketverkäufe abhängend von Niederschalgswerten und Temperatur

In einer ersten Analyse wird analysiert, wie viele Tickets an welchem Spieltag mit welchem Wetter verkauft wurden.

Zuerst werden alle verkauften Tickets gezählt. Da wir pro verkauftes Ticket einen Datensatz haben, wird eine Gruppierung mit den Felder Heimmannschaft, Gastmannschaft und Anpfiff gemacht. Dannach wird pro Datensatz die selbst geschriebene funktion fetch_weather_data() benützt, um die Wetterdaten beim Anpfiff zu laden. Es sollen verschiedene Plots anhand von diesen Daten erstellt werden. Zwei Liniendiagramme und zwei Scatterplots für Regen und Temperatur im vergleich zu den verkauften Tickets pro Spiel .

```python
# Initialisierung einer leere Liste, in die wir die Wetterdaten für jede Spielzeit einfügen
weather_data_list = []


selected_columns = ['Heimmannschaft', 'Gastmannschaft', 'Anpfiff']
result_table_spiele = fcsg_ticket_master_data[selected_columns].drop_duplicates().reset_index(drop=True)

# Zählen der Anzahl von Tickets pro Spiel
ticket_counts = fcsg_ticket_master_data.groupby(['Heimmannschaft', 'Gastmannschaft', 'Anpfiff']).size()

# Durch jede Zeile in fcsg_ticket_data loopen
for i in range(len(result_table_spiele)):
    # Extrahieren des spezifischen Datums und der Stunde
    match_case_date = result_table_spiele.at[i,'Anpfiff'].strftime('%Y-%m-%d')
    match_case_hour = result_table_spiele.at[i,'Anpfiff'].hour

    # Abrufen der Wetterdaten für das spezifische Datum
    df_weather_for_matchday = fetch_weather_data(match_case_date)

    # Filtern der Wetterdaten für die spezifische Stunde des Anpfiffs
    df_weather_for_matchhour = df_weather_for_matchday[df_weather_for_matchday['time'].dt.hour == match_case_hour].copy()

    # Hinzufügen der zusätzlichen Felder zu df_weather_for_matchhour
    df_weather_for_matchhour.loc[:, 'Heimmannschaft'] = result_table_spiele.at[i, 'Heimmannschaft']
    df_weather_for_matchhour.loc[:, 'Gastmannschaft'] = result_table_spiele.at[i, 'Gastmannschaft']
    df_weather_for_matchhour.loc[:, 'Anpfiff'] = result_table_spiele.at[i, 'Anpfiff']


    # Hinzufügen der Anzahl der verkauften Tickets
    df_weather_for_matchhour.loc[:, 'VerkaufteTickets'] = ticket_counts[(result_table_spiele.at[i, 'Heimmannschaft'], result_table_spiele.at[i, 'Gastmannschaft'], result_table_spiele.at[i, 'Anpfiff'])]

    # Hinzufügen der gefilterten Wetterdaten zu unserer Liste
    weather_data_list.append(df_weather_for_matchhour)

# Konvertieren der Liste der DataFrames in einen einzigen DataFrame
df_weather_for_all_matchdays = pd.concat(weather_data_list)

# Ändern der Reihenfolge der Spalten
cols_to_order = ['Heimmannschaft', 'Gastmannschaft', 'Anpfiff', 'VerkaufteTickets']
new_columns = cols_to_order + (df_weather_for_all_matchdays.columns.drop(cols_to_order).tolist())
df_weather_for_all_matchdays = df_weather_for_all_matchdays[new_columns]

# Anzeigen der gesammelten Wetterdaten
display(df_weather_for_all_matchdays)


```

```python

# Spiele als lesbare Zeichenkette erstellen und als Index setzen
df_weather_for_all_matchdays['Spiel'] = df_weather_for_all_matchdays['Anpfiff'].dt.strftime('%Y-%m-%d')
df_weather_for_all_matchdays.set_index('Spiel', inplace=True)

# Grösse des Plots festlegen
fig, ax1 = plt.subplots(figsize=(12, 8))

# Verkaufte Tickets auf der linken Y-Achse plotten
ax1.plot(df_weather_for_all_matchdays.index, df_weather_for_all_matchdays['VerkaufteTickets'], color='blue', label='Verkaufte Tickets')

# Legende für linke Y-Achse erstellen
ax1.legend(loc='upper left')
ax1.set_ylabel('Verkaufte Tickets')

# Zweite Y-Achse für den Niederschlag erstellen
ax2 = ax1.twinx()
ax2.plot(df_weather_for_all_matchdays.index, df_weather_for_all_matchdays['precipitation'], color='green', label='Niederschlag')

# Legende für rechte Y-Achse erstellen
ax2.legend(loc='upper right')
ax2.set_ylabel('Niederschlag (mm)')

# Titel und X-Achsen-Label hinzufügen
plt.title('Verkaufte Tickets und Niederschlag pro Spiel')
plt.xlabel('Spiel')

# X-Achse drehen für bessere Lesbarkeit und nur jeden fünften Tick anzeigen
xticks = ax1.xaxis.get_major_ticks()
for i in range(len(xticks)):
    if i % 5 != 0:
        xticks[i].label1.set_visible(False)
plt.xticks(rotation='vertical')

plt.show()
```

### 3.5.2 Analyse 2: Zeitlicher Ablauf der Ticketverkäufe abhängig vom Wochentag des Spiels

Als weitere Analyse wird der zeitliche Ablauf der Ticketverkäufe, sprich, wie viel Tage vor dem Spiel wie viele Tickets verkauft werden, vom jeweiligen Wochentag, an dem das Spiel stattfand, abhängt. Dazu wurden die Mittelwerte der Ticketverkäufe pro Wochentag pro Tag vor dem Spiel abgebildet.

Für die Analyse wurde ein Betrachtungszeitraum von 50 Tagen vor dem Spiel gewählt. Es ist aber zu beachten, dass Tickets bis zu einem Jahr vor dem Spiel gekauft werden können.

Insgesamt wurden 3 Grafiken erstellt.
```python
#Gruppierung der Masterdaten nach 'Heimmannschaft', 'Gastmannschaft', 'Anpfiff', 'Anpfiff Datum', 'Anpfiff Wochentag', 'Tage bis zum Spiel' und berechnung der Anzahl Karten
verkaufte_tickets_pro_spiel_und_tag = fcsg_ticket_master_data.groupby(['Heimmannschaft', 'Gastmannschaft', 'Anpfiff', 'Anpfiff Datum', 'Anpfiff Wochentag', 'Tage bis zum Spiel'])['Anzahl Karten'].size().reset_index()

#Gruppierung nach Wochentag und Tage bis zum Spiel und Berechnung des Mittelwertes
tickets_bis_spiel = verkaufte_tickets_pro_spiel_und_tag.groupby(['Anpfiff Wochentag', 'Tage bis zum Spiel'])['Anzahl Karten'].mean()

# Reihenfolge der Wochentage festlegen
wochentage_reihenfolge = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Gruppierung nach Wochentagen in die richtige Reihenfolge bringen
tickets_bis_spiel = tickets_bis_spiel.reindex(wochentage_reihenfolge, level=0)

# Filtere Spieltage zwischen 0 und 50
tickets_bis_spiel_filtered = tickets_bis_spiel[(tickets_bis_spiel.index.get_level_values('Tage bis zum Spiel') >= 0) & (tickets_bis_spiel.index.get_level_values('Tage bis zum Spiel') <= 50)]

# Reset Index
tickets_bis_spiel_filtered = tickets_bis_spiel_filtered.reset_index()

print(tickets_bis_spiel_filtered)
```

```python
# Analyse 1.1: Gruppierung nach Wochentag und Anpfiffszeit

# Mastertabelle mit richtigen Columns vorbereiten
selected_columns = ['Heimmannschaft', 'Gastmannschaft', 'Anpfiff Datum', 'Anpfiff Zeit', 'Anpfiff Wochentag', 'Gruppierung']
reduzierte_tabelle = fcsg_ticket_master_data[selected_columns].copy()

# Gruppierung der Tabelle mit Auswertung der Anzahl Ticketverkäufe pro Tag und Anpfiffszeit
ticket_counts_day = reduzierte_tabelle.groupby(['Heimmannschaft', 'Gastmannschaft', 'Anpfiff Datum', 'Anpfiff Zeit', 'Anpfiff Wochentag', 'Gruppierung']).size().reset_index(name='Anzahl Tickets')

# Gruppierung nach 'Wochentag' und 'Anpfiffszeit' und Berechnung des Mittelwerts
neue_gruppierung = ticket_counts_day.groupby(['Anpfiff Wochentag', 'Anpfiff Zeit'])['Anzahl Tickets'].mean()
# Reihenfolge der Wochentage festlegen
wochentage_reihenfolge = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# Gruppierung nach Wochentagen in die richtige Reihenfolge bringen
neue_gruppierung = neue_gruppierung.reindex(wochentage_reihenfolge, level=0)

#Diagrammgrösse
plt.figure(figsize=(10, 6))
# Balkendiagramm erstellen
neue_gruppierung.plot(kind='bar')
# Diagrammtitel und Achsenbeschriftungen hinzufügen
plt.title('Durchschnittliche Anzahl verkaufter Tickets pro Wochentag und Anpfiffszeit')
plt.xlabel('Wochentag, Anpfiffszeit')
plt.ylabel('Durchschnittliche Anzahl verkaufter Tickets')
# Diagramm anzeigen
plt.show()
```

```python
#Analyse 1.2: Gruppierung nach Wochentag
# Gruppierung nach 'Anpfiff Wochentag' und 'Anpfiff Zeit' und Berechnung des Mittelwerts
neue_gruppierung_wochentag = ticket_counts_day.groupby(['Anpfiff Wochentag', 'Gruppierung'])['Anzahl Tickets'].mean().unstack(fill_value=0)

# Reihenfolge der Wochentage festlegen
wochentage_reihenfolge = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Gruppierung nach Wochentagen in die richtige Reihenfolge bringen
neue_gruppierung_wochentag = neue_gruppierung_wochentag.reindex(wochentage_reihenfolge, level=0)

# Gestapeltes Balkendiagramm erstellen
ax = neue_gruppierung_wochentag.plot(kind='bar', stacked=True, figsize=(20, 10))

# Legende oben rechts anzeigen
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1))

# Diagrammtitel und Achsenbeschriftungen hinzufügen
plt.title('Durchschnittliche Anzahl verkaufter Tickets pro Wochentag der Spiele')
plt.xlabel('Wochentag')
plt.ylabel('Durchschnittliche Anzahl verkaufter Tickets')

# Diagramm anzeigen
plt.show()
```
