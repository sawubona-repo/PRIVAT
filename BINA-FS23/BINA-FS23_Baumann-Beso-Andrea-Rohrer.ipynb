#### Step 1.2.1 Leerwohnungsziffer
In der Leerwohnungszählung gelten (anders als bei der Gebäude- und Wohnungszählung) nur diejenigen Wohnungen als Leerwohnungen, die auf dem Markt zur Dauermiete bzw. zum Kauf angeboten werden. Es werden alle bewohnbaren Wohnungen berücksichtigt, unabhängig davon, ob sie möbliert oder nicht möbliert sind; mitgerechnet werden auch leerstehende Ferien- oder Zweitwohnungen, sofern sie das ganze Jahr bewohnbar und zur Dauermiete (mindestens 3 Monate) oder zum Verkauf ausgeschrieben sind.

Unter der Leerwohnungsziffer versteht man den prozentualen Anteil der leerstehenden Wohnungen (Stichtag: 1. Juni) am Gesamtwohnungsbestand der registerbasierten Gebäude- und Wohnungsstatistik (GWS) des Vorjahres.


#### Step 1.2.2 Negativzinsen
Die Einführung von Negativzinsen ist eine unkonventionelle geldpolitische Massnahme, die von Zentralbanken ergriffen wird, um die wirtschaftliche Aktivität anzukurbeln und die Inflation anzutreiben. Im Jahr 2014 führte die Schweizerische Nationalbank (SNB) erstmals Negativzinsen ein, um den Schweizer Franken abzuwerten und die Wirtschaft vor den Auswirkungen einer überbewerteten Währung zu schützen. Diese Massnahme war eine Reaktion auf den starken Zustrom von Kapital in die Schweiz aufgrund der globalen Unsicherheit und der Flucht in als sicher geltende Anlagen. Die Negativzinsen haben direkte und indirekte Auswirkungen auf die Mietpreise und das Wohnungsangebot in der Schweiz.

Direkt führen die Negativzinsen zu günstigeren Finanzierungsmöglichkeiten für Immobilieninvestoren und ermöglichen es ihnen, Kredite zu niedrigeren Kosten aufzunehmen. Dies kann zu einer erhöhten Nachfrage nach Immobilien führen, da Investoren attraktive Renditen suchen. Indirekt können Negativzinsen auch Auswirkungen auf die Vermietung von Wohnungen haben. Wenn Immobilieninvestoren günstige Finanzierungsmöglichkeiten nutzen, um neue Wohnungen zu bauen, könnte dies zu einem Anstieg des Wohnungsangebots führen.

Es können jedoch auch gegenteilige Entwicklungen entstehen, wenn Anleger vermehrt nach renditestarken Alternativen suchen, da herkömmliche Anlagen wie Sparkontos oder Anleihen geringe oder negative Renditen bieten. Infolgedessen könnte ein verstärkter Zustrom von Kapital in den Immobilienmarkt die Preise weiter in die Höhe treiben und zu einem Mangel an bezahlbarem Wohnraum führen.

```python
# Google Drive mounten
drive.mount('/content/drive')

# Pfad zur Excel-Datei auf Google Drive angeben
file_path_region = '/content/drive/MyDrive/bina/colab/bina2/DatensatzRegionen.xlsx'

# Excel-Datei einlesen
dfReg = pd.read_excel(file_path_region)

# Dataframe anzeigen (5 erste Datensätze)
dfReg.head()
```

```python
# Daten für das Diagramm auswählen
data = dfReg[['Jahr', 'Region', 'Leerwohnungsziffer']]

# Eindeutige Regionen identifizieren
regionen = data['Region'].unique()

print(data)
```

#### Step 4.1.3: Entwicklung der Leerwohnungsziffer je Region

```python
# Farben für jede Region definieren
farben = ['violet', 'blue', 'green', 'orange', 'purple', 'khaki', 'lightcoral']

# Grösse des Plots festlegen (10 Zoll (Breite) und 7 Zoll (Höhe))
fig, ax = plt.subplots(figsize=(10, 7))

# Liniendiagramm erstellen
for region, farbe in zip(regionen, farben):
    region_data = data[data['Region'] == region]
    ax.plot(region_data['Jahr'], region_data['Leerwohnungsziffer'], color=farbe, label=region)

# Durchschnitt der Leerwohnungsziffer über die Jahre berechnen
durchschnitt = data.groupby('Jahr')['Leerwohnungsziffer'].mean()

# Kurve für den Durchschnitt hinzufügen
ax.plot(durchschnitt.index, durchschnitt.values, linestyle='--', color='red', label='Durchschnitt')

# Achsenbeschriftungen und Titel hinzufügen
ax.set_xlabel('Jahr')
ax.set_ylabel('Leerwohnungsziffer')
ax.set_title('Entwicklung der Leerwohnungsziffer nach Regionen')

# Legende einblenden
ax.legend()

# Diagramm anzeigen
plt.show()
```

In der erstellten Grafik wird der Mittelwert mittels einer rot gestrichelten Linie dargestellt. Aus der Abbildung geht eine gleichmässige Entwicklung der Leerwohnungsziffer aller Regionen hervor. Dazu zählt der Rückgang der nicht ständigen Wohnbevölkerung, die Veränderung der Haushaltszusammensetzung, die geringe Neubautätigkeit und die Art der Neubautätigkeit.
Nachfolgend wird überprüft, inwiefern sich die Leerwohnungsziffer je Region über den gesamten Zeitraum von 2010 bis 2021 verändert hat. Also um wieviel sich die Leerwohnungsziffer von 2010 bis 2021 verändert hat.


```python
# Filtern der Daten von 2010 und 2021
df_2010 = dfReg[dfReg['Jahr'] == 2010]
df_2021 = dfReg[dfReg['Jahr'] == 2021]

# Berechnung der Veränderung der Leerwohnungsziffer
df_2010.set_index('Region', inplace=True)
df_2021.set_index('Region', inplace=True)
df_change = df_2021['Leerwohnungsziffer'] - df_2010['Leerwohnungsziffer']

# Berechnung des Mittelwerts
mean_change = df_change.mean()

df_change.head
```

```python
# Erstellen des Balkendiagramms
plt.figure(figsize=(10, 7))
bars = plt.bar(df_change.index, df_change.values)

# Hinzufügen der Balkensäule für den Mittelwert
plt.axhline(mean_change, color='red', linestyle='--', label='Mittelwert')

# Hinzufügen der Werte in den Balken
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2), ha='center', va='bottom')

# Hinzufügen von Achsentiteln und Diagrammtitel
plt.xlabel('Region')
plt.ylabel('Veränderung der Leerwohnungsziffer')
plt.title('Veränderung der Leerwohnungsziffer je Region (2010-2021)')

# Anzeigen der Legende
plt.legend()

# Anzeigen des Diagramms
plt.show()
```

Anhand dieser Abbildung ist ersichtlich, um wieviel Prozentpunkte die Leerwohnungsziffer im Jahr 2021 die Werte des Jahres 2010 übersteigen.

Durchschnittlich hat die Leerwohnungsziffer um 0.74 Pozentpunkte zugelegt. Erneut sticht die Region bzw. der Kanton Tessin hervor, in welcher eine Zunahme von 2.15 Prozentpunkten zu verzeichnen ist. Dieser Ausreisser wurde entsprechend überprüft. Dazu wurde eine weitere Abbildung erstellt, welche die Anzahl leerstehender Wohnungen über den gesamten Zeitraum für die Region Tessin darstellt.

```python
# Filtern der Daten für die Region Tessin
df_tessin = dfReg[dfReg['Region'] == 'Tessin']

df_tessin.head()
```

Im Zeitraum 2010 bis 2021 hat sich die Anzahl leerstehender Wohnungen im Tessin kontinuierlich erhöht. Im Jahr 2010 standen 1458 Wohnungen leer, wohingegen im Jahr 2021 7017 Wohnungen leer standen. Dies entspricht einer Zunahme von 381%.

Gemäss Swissinfo ([2019](https://www.swissinfo.ch/ger/alle-news-in-kuerze/leerstandsquote-am-mietwohnungsmarkt-steigt---tessin-in-schieflage/45257222)) ist diese Veränderung im Tessin auf mehrere Faktoren zurückzuführen. In den Jahren 2012 bis 2015 gab es eine Phase überdurchschnittliche wirtschaftlicher Dynamik, die zu kräftigen Investitionen in den Mietwohnungsbau führte. Der Scatter Plot oder auch Korrelations-Analyse genannt, wird verwendet, um den Zusammenhang zwischen zwei Variablen darzustellen. Auf der Y-Achse ist die Variable «Durchschnittlicher Mietpreis» und auf der X-Achse der Wert «Leerwohnungsziffer» dargestellt. Der Scatter Plot ermöglicht es, einen Zusammenhang zwischen diesen beiden Variablen zu analysieren. Die einzelnen Werte sind entlang der Y- und X-Achse verteilt. Jeder einzelne Punkt repräsentiert einen bestimmten Zeitpunkt im Jahr und den entsprechenden durchschnittlichen Preis und die Leerwohnungsziffer. Die Linie, welche die Punkte verbindet, zeigt den Verlauf oder den Trend der Mietpreise und der Leerwohnungsziffer im Laufe der Jahre an. Diese Darstellung ermöglicht es uns, verschiedene Muster und Trends über die Jahre zu erkennen.

Wenn die Punkte entlang der Linie nach oben tendieren, deutet dies auf eine positive Korrelation hin, was bedeutet, dass höhere Leerwohnungsziffern mit höheren durchschnittlichen Mietpreisen einhergehen. Wenn die Punkte entlang der Linie nach unten tendieren, zeigt dies eine negative Korrelation an, was bedeutet, dass höhere Leerwohnungsziffern mit niedrigeren durchschnittlichen Mietpreisen verbunden sind.

Aus der Abbildung ist zu sehen, dass in allen Regionen ein positiver Korrelationskoeffizient zu erkennen ist. Alle Regionen zeigen eine schwache positive Korrelation auf, ausser die Region Tessin, welche eine starke positive Korrelation aufweist. Zusammenfassend ist zu sagen, dass höhere Mietpreise tendenziell mit höheren Leerwohnungsziffern einhergehen.

Der Scatter Plot ist zwar für eine visuelle Darstellung des Zusammenhangs zwischen den Variablen nützlich, jedoch bestätigt er keine Kausalität. Unsere Analysen deuten jedoch darauf hin, dass eine Kausalität zwischen dem durchschnittlichen Mietwohnungspreis und der Leerwohnungsziffer besteht. Weitere Faktoren müssten berücksichtigt werden, um eine Kausalität zwischen diesen beiden Variablen bestätigen zu können.

Zudem ist auch zu beachten, dass in den Grossstädten wie Zürich, Zug, Genf oder auch Basel immer mehr Menschen leben und somit auch immer mehr Wohnungen benötigt werden. Hinzu kommen aussenpolitische Konflikte wodurch die Wohnungssituation weiter Unterdruck gerät. Generell wird festgestellt, dass die Mietwohnungspreise kontinuierlich ansteigen während die Leerwohnungsbestände zurückgehen.


Für die kommenden Jahren ist noch keine Verbesserung der Situation in Sicht. Insbesondere mit der anhaltenden Inflation kann man damit rechnen, dass die Mietwohnungspreise in den kommenden Jahren weiter ansteigen. Zudem wurde erst kürzlich kommuniziert, dass der Referenzzinssatz per Juni 2023 um 0.25 Prozentpunkte auf neu 1.5% ansteigt. Auch mit dieser Erhöhung werden die Mietzinse weiterhin steigen. Trotz sinkender Leerstände und deutlich steigenden Mietpreisen planen Investoren derzeit keine Ausweitung der Wohnbautätigkeit. Die steigenden Baupreise, erhöhte Finanzierungskosten, höhere administrativen Hürden sowie deutlich gestiegene Opportunitätskosten dämpfen zurzeit die Bautätigkeiten. Für die Investoren ist die heutige Sicht nicht sehr attraktiv (Tages Anzeiger, 2023):  Eine Besserung ist bis auf Weiteres noch nicht in Sicht.
