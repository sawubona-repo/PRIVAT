# 4.1.1 Bruttoinlandsprodukt Deutschland
Das Bruttoinlandsprodukt (BIP) ist eine ökonomische Kennzahl, welche den Gesamtwert alle in einer Volkswirtschaft produzierten Waren und Dienstleistungen innerhalb eines bestimmten Zeitraums misst. Es dient als Mass für die wirtschaftliche Aktivität und den Wohlstand eines Landes (Storp, 2022).

```python
# Diagramm für BIP pro Jahr
plt.plot(bip['Jahr'], bip['BIP'], label="BIP in Mio. EUR")

# Trendlinie einzeichnen
z = np.polyfit(bip['Jahr'], bip['BIP'], 1)
p = np.poly1d(z)
plt.plot(bip['Jahr'], p(bip['Jahr']), color='red', linestyle='--', label="Trendlinie", alpha=0.5)

plt.xlabel=("Jahr")
plt.title('BIP Deutschland')
plt.legend()
plt.show()

# Abweichung berechnen
abweichung = bip['BIP'] - p(bip['Jahr'])
abweichungs_df = pd.DataFrame({'Jahr': bip['Jahr'], 'Abweichung': abweichung})

# Diagramm für Abweichung zur Trendlinie pro Jahr
abweichungs_df.plot(x='Jahr', y='Abweichung', kind='line', label="Abweichung in Mio. EUR")
plt.axhline(0, color='red', linestyle='-', label="0-Wert", alpha=0.25)

plt.xlabel=('Jahr')
plt.title('Abweichung zur Trendlinie')
plt.legend()
plt.show()
```

Anhand der Diagramme ist zu erkennen, dass es sowohl Zeiträume gibt, welche unterhalb als auch oberhalb der Trendlinie liegen. Dennoch besteht ein kontinuierlicher Anstieg des BIPs.

Der rasante Anstieg des BIPs vom Jahr 1990 auf 1991 und folgend, kann seine Ursache in der Wiedervereinigung Deutschland haben. Die regierende Partei zu dieser Zeit war unter anderem die CDU. Somit könnte dieser Partei ein positiver Einfluss auf den BIP Deutschlands und somit der Wirtschaft zugeschrieben werden.

Ebenfalls ersichtlich ist, dass in 2023 erstmalig die deutsche Inflationsrate höher ist als die weltweite. Die starken Ausreisser innerhalb der weltweiten Inflationsrate zwischen 1990 und 1996 haben unter anderem den Grund in dem Zusammenbruch des Ostblocks, Nachwirkungen der Öl-Krise und der Währungs- und Finanzkrise. Die stabile Inflationsrate Deutschlands zwischen 1990 und 2022 lässt auf eine solide und starke Wirtschaft und somit auch auf eine gute Politik schliessen.
```python
# Differenz berechnen und Diagramm erstellen
dif = inflationWeltweit['Inflationsrate'] - inflationD['Inflationsrate']
plt.plot(inflationWeltweit['Jahr'], dif, color='black', label="Differenz W:D")
plt.axhline(y=0, color='blue', linestyle='-', alpha=0.25, label="0-Wert")
plt.axhline(y=dif.mean(), color='green', linestyle='--', alpha=0.25, label="⌀ Differenz")

# Trendlinie einzeichnen
z = np.polyfit(inflationWeltweit['Jahr'], dif, 1)
p = np.poly1d(z)
plt.plot(inflationWeltweit['Jahr'], p(inflationWeltweit['Jahr']), color='red', linestyle='--', label="Trendlinie", alpha=0.5)

plt.xlabel=("Jahr")
plt.ylabel=("Differenz")
plt.title("Differenz Inflationsrate W:D")
plt.legend()
plt.show()
```

Das Diagramm zeigt die Differenz der deutschen Inflationsrate im Vergleich zur weltweiten. Aufgrund der Ausreisser zwischen 1990 und 1996 werden die Trendlinie und die durchschnittliche Differenz stark beeinflusst und verfälscht. Deshalb werden im nächsten Schritt diese Ausreisser entfernt und der Zeitraum von 1997 bis 2023 genauer betrachtet. Diese Eliminierung der Ausreisser sorgt für einen besseren Überblick über die Entwicklung der Differenz der beiden Raten und sorgt evtl. für weitere Erkenntnisse.

```python
# Betrachtungszeitraum 1997 - 2023 (Eliminierung der Ausreisser)
inflationWeltFiltered = inflationWeltweit[(inflationWeltweit['Jahr'] >= 1997) & (inflationWeltweit['Jahr'] <= 2023)]
inflationDFiltered = inflationD[(inflationD['Jahr'] >= 1997) & (inflationD['Jahr'] <= 2023)]

# Differenz berechnen und Diagramm erstellen
dif = inflationWeltFiltered['Inflationsrate'] - inflationDFiltered['Inflationsrate']
plt.plot(inflationWeltFiltered['Jahr'], dif, color='black', label="Differenz W:D")
plt.axhline(y=0, color='blue', linestyle='-', alpha=0.25, label="0-Wert")
plt.axhline(y=dif.mean(), color='green', linestyle='--', alpha=0.25, label="⌀ Differenz")

# Trendlinie einzeichnen
z = np.polyfit(inflationWeltFiltered['Jahr'], dif, 1)
p = np.poly1d(z)
plt.plot(inflationWeltFiltered['Jahr'], p(inflationWeltFiltered['Jahr']), color='red', linestyle='--', label="Trendlinie", alpha=0.5)

plt.xlabel=("Jahr")
plt.ylabel=("Differenz")
plt.title("Differenz Inflationsrate 1997 - 2023")
plt.legend()
plt.show()
```

Nach dem Aussortieren der Werte, welche einen Ausschlag darstellten, ist nun deutlich zu erkennen, dass die Trendlinie einen negativen Trend aufweist. Dies bedeutet, dass die deutsche Inflationsrate näher an die weltweite rückt. Das kann ein Indiz für eine Verschlechterung der Stellung Deutschlands im Weltmarkt sein.

Aufgrund der Krisen, wie der Ukraine-Krieg oder die COVID-Pandemie, steigen beide Raten an. Ein Grund hierfür ist unter anderem die Klima- und Wirtschaftspolitik der SPD, Grünen und FDP welche zu diesem Zeitpunkt regieren.

Somit kann als Fazit aus dieser Betrachtung entnommen werden, dass die SPD, Grünen und FPD einen schlechten Einfluss auf die deutsche Inflationsrate haben. Jedoch ist diese Aussage mit Vorsicht zu geniessen, da bei dieser Betrachtung nicht alle Faktoren und Einflüsse berücksichtig wurden.

# 4.2.2 Auszubildende
Ein weiterer wichtiger Aspekt ist die Anzahl an Auszubildenden. Aufgrund eines enormen Mangels an Auszubildenden ist dieses Thema in der heutigen Zeit immer öfters vertreten und wird kritisch angesehen. Im Folgenden wird überprüft, wie sich die Anzahl der Auszubildenden im Verhältnis zur Gesamteinwohnerzahl Deutschlands verändert hat.

```python
# Diagramm erstellen
fig, ax1 = plt.subplots()

#Graph für Einwohner
ax1.plot(einwohnerD['Jahr'], einwohnerD['Einwohner'], color='blue', label="Einwohner in D")
ax1.set_ylabel("Einwohner in Mio.", color='blue')
plt.legend(loc='upper left')

# Graph für Auszubildende
ax2 = ax1.twinx()
ax2.plot(auszubildende['Jahr'], auszubildende['Anzahl'], color='green', label="Auszubildende")
ax2.set_ylabel("Auszubildende in 100'000", color='green')
plt.legend(loc='upper right')

# Trendlinie für Einwohner berechnen
x = einwohnerD['Jahr']
y = einwohnerD['Einwohner']
coefficients = np.polyfit(x, y, 3)  # Lineare Regression, Grad 3
trendline = np.polyval(coefficients, x)

# Trendlinie für Auszubildende berechnen
x_auszubildende = auszubildende['Jahr']
y_auszubildende = auszubildende['Anzahl']
coefficients_auszubildende = np.polyfit(x_auszubildende, y_auszubildende, 2)  # Lineare Regression, Grad 2
trendline_auszubildende = np.polyval(coefficients_auszubildende, x_auszubildende)

# Trendlinie in das Diagramm einzeichnen
ax1.plot(x, trendline, color='blue', label='Trendlinie Einwohner', alpha=0.25)
ax2.plot(x_auszubildende, trendline_auszubildende, color='green', label='Trendlinie Auszubildende', alpha=0.25)

ax1.legend(loc="upper left")
ax1.set_xlabel("Jahr")
ax1.set_ylim(79.5, 85)
ax2.set_ylim(1200000, 1800000)
plt.legend()
plt.show()

# Gleiche Länge beider Trendlinien
resampled_trendline_auszubildende = resample(trendline_auszubildende, len(trendline))

# Korrelationskoeffizient berechnen
correlation = np.corrcoef(trendline, resampled_trendline_auszubildende)[0, 1]
print("Korrelationskoeffizient : ",correlation)
```

Trotz der steigenden Bevölkerung in Deutschland, sinkt die Anzahl an Auszubildenden seit 2008 stark ab. Dies war jedoch nicht immer der Fall. Zwischen 1990 und 2011 verlaufen beide Graphen in einer positiven Korrelation ähnlich zueinander. Ab 2012 verlaufen beide Graphen in einer negativen Korrelation zueiander.

Idealerweise bedeutet eine Zunahme an Einwohnern eine ebenfalls steigende Anzahl an Auszubildenden. In diesem Beispiel ist dies jedoch nicht der Fall. Hierfür kann es viele verschiedene Gründe geben, jedoch sollte die Politik ein solches Vorkommnis möglichst verhindern. Ein Mangel an Auszubildenden ist heute nicht den fehlenden Ausbildungsplätzen zuzuschreiben, sondern dem Mangel an Personen, die eine Ausbildung anstreben. Für eine funktionstüchtige Wirtschaft und Gesellschaft benötigt es bestimmte Berufsgruppen, vor allem im Handwerk. Die Gesetzgebung und Strafverfolgungs- und Sicherheitspolitik kann sich sowohl positiv als auch negativ auf die Kriminalität im Land auswirken. Im Folgenden wird untersucht, ob sich Zusammenhänge zwischen der Politik und der Kriminalität erkennen lassen.

```python
# Diagramm erstellen
fig, ax1 = plt.subplots()

# Graph für Kriminalitaet
ax1.plot(kriminalitaet['Jahr'], kriminalitaet[' Straftaten'], color='blue', label="Kriminalität")
ax1.set_ylabel("Straftaten in Mio.", color='blue')
plt.legend(loc='lower right')

# Senkrechte Linien und Beschriftung einfügen
plt.axvline(x=1990, color='red', linestyle='--', alpha=0.25)
plt.axvline(x=1998, color='red', linestyle='--', alpha=0.25)
plt.axvline(x=2006, color='red', linestyle='--', alpha=0.25)
plt.axvline(x=2022, color='red', linestyle='--', alpha=0.25)
plt.annotate('CDU - FDP', xy=(1991, 6900000), xytext=(1991, 6900000))
plt.annotate('SPD - Grüne', xy=(1999.5, 6900000), xytext=(1999.5, 6900000))
plt.annotate('CDU - SPD/FDP', xy=(2011.5, 6900000), xytext=(2011.5, 6900000))

# Trendlinie für Kriminaliät
x = kriminalitaet['Jahr']
y = kriminalitaet[' Straftaten']
coefficients = np.polyfit(x, y, 3)  # Lineare Regression, Grad 3
trendline = np.polyval(coefficients, x)
ax1.plot(x, trendline, color='blue', label='Trendlinie Einwohner', alpha=0.25)

ax1.set_ylim(4500000, 7000000)
ax1.set_xlabel("Jahr")
plt.show()
```

Die Kriminalität steigt nach 1990 rasant an. Gründe hierfür sind die Wiedervereinigung, demographische Veränderungen durch die Ostblock-Auflösung und der Anstieg von organisierter Kriminalität. Ausserdem ist ab 2004 ein Abwärtstrend in der Anzahl an Straftaten zu erkennen. Dies deutet auf eine gute Kriminalitätspolitik der CDU hin.

Im nächsten Schritt wird überprüft, ob ein Zusammenhang zwischen der Anzahl an Straftaten und der Arbeitslosigkeit besteht.

```python
# Diagramm erstellen
fig, ax1 = plt.subplots()

# Graph für Kriminalität
ax1.plot(kriminalitaet['Jahr'], kriminalitaet[' Straftaten'], color='blue', label="Straftaten")
ax1.set_xlabel("Jahr")
ax1.set_ylabel("Straftaten in Mio.", color='blue')
plt.legend(loc='upper left')

# Trendlinie für Kriminaliät
x1 = kriminalitaet['Jahr']
y1 = kriminalitaet[' Straftaten']
coefficients1 = np.polyfit(x1, y1, 3)  # Lineare Regression, Grad 3
trendline1 = np.polyval(coefficients1, x1)
ax1.plot(x1, trendline1, color='blue', linestyle='--', label='Trendlinie Straftaten', alpha=0.5)
plt.legend(loc='upper left')

# Graph für Arbeitslosenquote
ax2 = ax1.twinx()
ax2.plot(arbeitslosenzahl['Jahr'], arbeitslosenzahl['Arbeitslosenquote'], color='green', label="Arbeitslosenquote")
ax2.set_ylabel("Arbeitslosenquote in %", color='green')

# Trendlinie für Arbeitslosenquote
x2 = arbeitslosenzahl['Jahr']
y2 = arbeitslosenzahl['Arbeitslosenquote']
coefficients2 = np.polyfit(x2, y2, 3)  # Lineare Regression, Grad 3
trendline2 = np.polyval(coefficients2, x2)
ax2.plot(x2, trendline2, color='green', linestyle='--', label='Trendlinie Arbeitslosenquote', alpha=0.5)

ax1.set_ylim(4500000, 7200000)
ax2.set_ylim(5.0, 12.0)
plt.legend()
plt.show()

# Korrelationskoeffizient berechnen
correlation = np.corrcoef(coefficients1, coefficients2)[0, 1]
print("Korrelationskoeffizient: ",correlation)
```

Das Diagramm zeigt deutlich, dass eine positive Korrelation zwischen der Arbeitslosigkeit und der Kriminalität besteht. Die Trendlinien der beiden Graphen weisen einen nahezu identischen Verlauf auf. Der berechnete Korrelationskoeffizient von 0.99999... belegt dies. Der Verlauf der beiden Trendlinien weist auf einen Zusammenhang hin, dass eine höhere Arbeitslosigkeit für mehr Kriminalität im Land sorgt. Eine hohe Arbeitslosigkeit bringt viele soziale und wirtschaftliche Aspekte mit sich, welche unter anderem dafür sorgen, dass die Kriminalität ebenfalls ansteigt.

Folglich kann der Schluss gezogen werden, dass anhand einer guten Politik der Parteien, die sich auf die Bekämpfung der Arbeitslosigkeit ebenfalls eine Senkung der Kriminalität erreicht werden kann. Anhand der ausgewerteten Daten sorgt die CDU für eine sinkende Arbeitslosigkeit und gleichzeitig einen Zurückgang der Kriminalität im Land.
